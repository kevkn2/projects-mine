<h1>Exploring the Steam Gaming Store Data</h1>

For this project, we will use the dataset from [here](https://www.kaggle.com/nikdavis/steam-store-games#steam.csv). Using data gathered from the Steam Store and SteamSpy APIs, this dataset provides information about various aspects of games on the store, such as its genre and the estimated number of owners.

Gathered around May 2019, it contains most games on the store released prior to that date. Unreleased titles were removed as well as many non-games like software, though some may have slipped through.

Here is the columns from the dataset:
- appid : Unique identifier for each title
- name : Title of app (game)
- release_date : Release date in format YYYY-MM-DD
- english : Language support: 1 if is in English
- developer : Name (or names) of developer(s). Semicolon delimited if multiple
- publisher : Name (or names) of publisher(s). Semicolon delimited if multiple
- platforms : Semicolon delimited list of supported platforms. At most includes: windows;mac;linux
- required_age : Minimum required age according to PEGI UK standards. Many with 0 are unrated or unsupplied.
- categories : Semicolon delimited list of game categories, e.g. single-player;multi-player
- genres : Semicolon delimited list of game genres, e.g. action;adventure
- steamspy_tags : Semicolon delimited list of top steamspy game tags, similar to genres but community voted, e.g. action;adventure
- achievements : Number of in-games achievements, if any
- positive_ratings : Number of positive ratings, from SteamSpy
- negative_ratings : Number of negative ratings, from SteamSpy
- average_playtime : Average user playtime, from SteamSpy
- median_playtime : Median user playtime, from SteamSpy
- owners : Estimated number of owners. Contains lower and upper bound (like 20000-50000). May wish to take mid-point or lower bound. Included both to give options.
- price : Current full price of title in GBP, (pounds sterling)

<h2>Dividing the Games Genre</h2>

To make our projects easier, first of all, we should see the genre column first, because there could be more than 1 genre from a single game. To begin this cleaning, we will see the unique values from the column.

```python
steam["genres"].value_counts(dropna = False)
```

After looking at the columns, we know that there are lots of genre in a game, which most of them are Indie, Action, and Adventure genre. According to this genre, we will:

1. Separate the Indie, Action, and Adventure genre from the rest of the column, and create new columns named with the genre, which gives True statement if the row has it
2. Erase the row if the game didn't have the genre, so we can focusing the project on these genres.

```python
#create the function that returns True if the game has that genre
def define_genre(genre, value):
    if genre in value:
        return True
    return False

#apply the function on the new columns
steam["action"] = steam["genres"].apply(define_genre, args = ("Action",))
steam["indie"] = steam["genres"].apply(define_genre, args = ("Indie",))
steam["adventure"] = steam["genres"].apply(define_genre, args = ("Adventure",))

#drop the row if the game didn't have the genres
steam = steam.drop(steam[(steam["action"] == False)
                        & (steam["indie"] == False)
                        & (steam["adventure"] == False)].index)

#let's see the the columns value
steam["action"].value_counts()
```

<h2>Create the Positive Percentage Column</h2>

As the column named, positive percentage column is the percentage of the positive ratings on the total ratings (positive rating + negative rating). Let's create the column.

```python
steam["positive_percent"] = steam["positive_ratings"]/(steam["positive_ratings"]+steam["negative_ratings"])*100
steam["positive_percent"].value_counts()
```

<h2>Searching the Most Played and the Most Positive Rating Games on Each Genre</h2>

To do this, we will:
1. Separate the games into different dataframes
2. Create bar plots that only shows the top 5 most played and most positive rating games in each dataframes

```python
#we will focus on doing on each genre
action = steam[steam["action"] == True]
action.head()
```

```python
#make function that create a top 5 list
def top_5(column, genre):
    top_5_list = np.sort(genre[column].unique())[::-1][:5]
    top_5_dataframe = genre[(genre[column] == top_5_list.tolist()[0])
                    |(genre[column] == top_5_list.tolist()[1])
                    |(genre[column] == top_5_list.tolist()[2])
                    |(genre[column] == top_5_list.tolist()[3])
                    |(genre[column] == top_5_list.tolist()[4])]
    return top_5_dataframe.sort_values(ascending = False,by = [column])[:5]
```

```python
#getting the top 5 most played action games
action_top5_mostplayed = top_5("average_playtime", action)
style = {"axes.spines.right" : False, #removing the right and top spines
        "axes.spines.top" : False}

#view the results on a bar plot
import matplotlib as plt
%matplotlib inline
table = action_top5_mostplayed.pivot_table(values = "average_playtime", index = "name")
table = table.reindex(table["average_playtime"].sort_values().index)#it's not sorted, so we need to sort the value
with plt.style.context(style):
    ax = table.plot(kind = "bar", legend = False, title = "Top 5 Most Played Action Games on Steam", yticks = [0, 10000, 20000])
    ax.set_yticklabels([0, "10k", "20k"])
```

As we can see, the most played game of all time was the original Counter-Strike, which is no suprise because it's everybody's favourite game, followed by Dead or Alive 5, Rainbow Six Siege, Counter-Strike: Source (the sequel from the first in this list) and Monster Hunter.

```python
#now, lets get to the top 5 best action games
action_top5_best = top_5("positive_percent", action)

#view the results on a bar plot
table = action_top5_best.pivot_table(values = "positive_percent", index = "name")
table = table.reindex(table["positive_percent"].sort_values().index)#it's not sorted, so we need to sort the value
with plt.style.context(style):
    ax = table.plot(kind = "bar", legend = False, title = "Top 5 Best Action Games in Steam", yticks = [0, 50, 100])
```
